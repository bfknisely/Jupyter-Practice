#### How far does this go?

Jupyter allows you to create a formatted "notebook" to intersperse code with markdown fields.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')

print('This is an example of a graph in a Jupyter notebook.')
```

Markdowns fields can be used to explain concepts, make headings, and generally format your notebook.

Style and equations from $LaTeX$ are also supported. 

For example, you could write the force balance for a vibrations problem:

$m \ddot{x} + c \dot{x} + k x = f(t)$
